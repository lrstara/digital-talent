### 1. Deskripsi Data

Tabel diatas berupa data produksi padi di tiap provinsi di Indonesia mulai dari tahun 1993 hingga 2013. Luas lahan dinyatakan dalam satuan hektar, produksi dinyatakan dalam ton, dan produktivitas dinyatakan dalam ton/hektar. Latitude dan longitude merupakan koordinat dari lahan pertanian padi.

### 2. Statistik Data

```python
data_padi.count()
```

```python
# Total 
sm = data_padi.luas_panen.sum()
print('- Total luas panen: ', sm, 'Ha')
sm2 = data_padi.produksi.sum()
print('- Total produksi padi: ', sm2 , 'Ton')

# Nilai maksimum
maks1 = data_padi.luas_panen.max()
maks2 = data_padi.produktivitas.max()
maks3 = data_padi.produksi.max()
print('- Luas lahan padi terbesar dari tahun 1993 hingga 2013 ialah', maks1, 'Ha. \n  Produksi padi tertinggi dari tahun 1993 hingga 2013 adalah', maks3, 'Ton. \n  Produktivitas tertinggi yaitu sebesar', maks2, '%')

# Nilai minimum
min1 = data_padi.luas_panen.min()
min2 = data_padi.produktivitas.min()
min3 = data_padi.produksi.min()
print('- Luas lahan padi terkecil dari tahun 1993 hingga 2013 ialah', min1, 'Ha. \n  Produksi padi terendah dari tahun 1993 hingga 2013 adalah', min3, 'Ton. \n  Produktivitas terendah yaitu sebesar', min2, '%')
```

### 3. Visualisasi Sederhana dan Narasi

```python
ax = data_padi.groupby('nama_provinsi').mean()['produktivitas']
ax = ax.plot(kind='bar', title ="Rata-rata Produktivitas padi per provinsi tahun 1993-2013", legend=True, fontsize=8)
ax.set_xlabel("Provinsi", fontsize=12)
ax.set_ylabel("Produktivitas(%)", fontsize=12)
```

```python
print('- Provinsi yang memiliki tingkat rata-rata produktivitas padi tertinggi dari tahun 1993-2013 adalah Provinsi Bali dengan  \n  nilai 55,12%.')
print('- Provinsi yang memiliki tingkat rata-rata produktivitas padi dari tahun 1993-2013 adalah Provinsi Kepulauan Riau \n  dengan nilai 13,37%.')
print('- Tiga provinsi yang memiliki rata-rata produktivitas tertinggi dari tahun 1993-2013 adalah Provinsi Bali, Provinsi Jawa \n  Timur, dan Provinsi Jawa Tengah.')
print('- Tiga provinsi yang memiliki rata-rata produktivitas terendah dari tahun 1993-2013 adalah Provinsi Kepulauan Riau, Provinsi \n  Kalimantan Timur, dan Provinsi Irian Barat.')
print('- Diperlukan penerapan inovasi teknologi lebih lanjut dalam usahatani padi di Provinsi yang masih memiliki rata-rata \n  produktivitas rendah.')
```

```python
ax = data_padi.groupby('nama_provinsi').mean()['luas_panen']
ax = ax.plot(kind='bar', title ="Rata-rata Luas Lahan Padi per Provinsi Tahun 1993-2013", legend=True, fontsize=12)
ax.set_xlabel("Provinsi", fontsize=12)
ax.set_ylabel("Luas Lahan (Ha)", fontsize=12)
```

```python
print('- Luas rata-rata lahan padi tertinggi pada tahun 1993-2013 adalah Provinsi Jawa Timur dan luas rata-rata lahan padi terendah\n  pada tahun 1993-2013 adalah Provinsi Kepulauan Riau.')
print('- Tiga provinsi yang memiliki rata-rata luas lahan padi tertinggi dari tahun 1993-2013 adalah Provinsi Jawa timur, Provinsi \n  Jawa Barat, dan Provinsi Jawa Tengah.')
print('- Tiga provinsi yang memiliki rata-rata luas lahan padi terendah dari tahun 1993-2013 adalah Provinsi Kepulauan Riau,\n  Provinsi B.Belitung, dan Provinsi DKI Jakarta.')
```

```python
ax = data_padi.groupby('tahun').mean()['produktivitas']
ax = ax.plot(kind='bar', title ="Rata-rata Produktivitas Padi Berdasarkan Tahun", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Produktivitas (%)", fontsize=12)
```

```python
print('- Dari tahun 1993 hingga tahun 2013, rata-rata produktivitas padi cenderung meningkat terutama pada 5 tahun terakhir. \n  Hal tersebut didorong dengan banyaknya inovasi teknologi yang diterapkan pada tingkat petani.')
print('- Dari data di atas, rata-rata produktivitas dari Prov. Kepulauan Riau cenderung rendah sejalan dengan rata-rata luasan \n  lahan yang kecil. Namun, besarnya rata-rata luas lahan padi untuk Provinsi Jawa Timur tidak menunjukkan adanya rata-rata\n  produktivitas padi yang optimal.')
print('- Diperlukan penerapan teknologi pertanian di tingkat petani, seperti sistem pemupukan; bibit padi berkualitas; sistem \n  pengairan; dan sistem usahatani yang disesuaikan dengan kondisi lahan dan iklim di masing-masing Provinsi secara menyeluruh,\n  agar produktivitas padi dapat lebih dioptimalkan.')
```
