## 3. Visualisasi Sederhana dan Narasi

### Rata Produktivitas per Provinsi

```python
ax = data_padi.groupby('nama_provinsi').mean()['produktivitas']
ax = ax.plot(kind='bar', title ="Rata-rata Produktivitas padi per provinsi tahun 1993-2013", legend=True, fontsize=8)
ax.set_xlabel("Provinsi", fontsize=12)
ax.set_ylabel("Produktivitas(%)", fontsize=12)
```

   Provinsi yang memiliki tingkat rata-rata produktivitas padi tertinggi dari tahun 1993-2013 adalah Provinsi Bali dengan nilai 55,12%. Provinsi yang memiliki tingkat rata-rata produktivitas padi dari tahun 1993-2013 adalah Provinsi Kepulauan Riau  dengan nilai 13,37%. Tiga provinsi yang memiliki rata-rata produktivitas tertinggi dari tahun 1993-2013 adalah Provinsi Bali, Provinsi Jawa Timur, dan Provinsi Jawa Tengah. Tiga provinsi yang memiliki rata-rata produktivitas terendah dari tahun 1993-2013 adalah Provinsi Kepulauan Riau, Provinsi Kalimantan Timur, dan Provinsi Irian Barat. Diperlukan penerapan inovasi teknologi lebih lanjut dalam usahatani padi di Provinsi yang masih memiliki rata-rata produktivitas rendah.

### Rata Luas Lahan Padi per Provinsi

```python
ax = data_padi.groupby('nama_provinsi').mean()['luas_panen']
ax = ax.plot(kind='bar', title ="Rata-rata Luas Lahan Padi per Provinsi Tahun 1993-2013", legend=True, fontsize=8)
ax.set_xlabel("Provinsi", fontsize=8)
ax.set_ylabel("Luas Lahan (Ha)", fontsize=12)
```

Luas rata-rata lahan padi tertinggi pada tahun 1993-2013 adalah Provinsi Jawa Timur dan luas rata-rata lahan padi terendah pada tahun 1993-2013 adalah Provinsi Kepulauan Riau. Tiga provinsi yang memiliki rata-rata luas lahan padi tertinggi dari tahun 1993-2013 adalah Provinsi Jawa timur, Provinsi Jawa Barat, dan Provinsi Jawa Tengah. Tiga provinsi yang memiliki rata-rata luas lahan padi terendah dari tahun 1993-2013 adalah Provinsi Kepulauan Riau, Provinsi B.Belitung, dan Provinsi DKI Jakarta.

### Rata Produktivitas Padi per Tahun

```python
ax = data_padi.groupby('tahun').mean()['produksi']
ax = ax.plot(kind='line', title ="Rata-rata Produktivitas Padi", legend=True, fontsize=11)
```

Terlihat dari grafik di atas bahwa sejak tahun 1993 hingga tahun 2013, rata-rata produktivitas padi Indonesia cenderung meningkat walaupun terdapat beberapa periode yang mengalami penurunan. Kecenderungan dalam meningkatnya produktivitas padi di Indonesia didorong dengan banyaknya inovasi teknologi yang diterapkan pada tingkat petani. Pada data sebelumnya, diketahui bahwa rata-rata produktivitas dari Provinsi Kepulauan Riau cenderung rendah dan hal itu sejalan dengan rata-rata luasan lahan yang kecil. Hal tersebut juga berlaku pada provinsi Jawa Timur di mana luas lahan pada provinsi tersebut memang tinggi sehingga produktivitasnya juga meningkat. Namun, belum diketahui berapa pengaruh luas lahan terhadap produktivitas padi sehingga dibutuhkan analisis lebih lanjut. 

Peningkatan produktivitas padi dapat dilakukan dengan melakukan penerapan teknologi pertanian di tingkat petani, seperti sistem pemupukan; bibit padi berkualitas; sistem pengairan; dan sistem usahatani yang disesuaikan dengan kondisi lahan dan iklim di masing-masing Provinsi secara menyeluruh. Untuk ekstensifikasi lahan padi belum dapat dilakukan mengingat pembukaan lahan baru untuk pertanian cukup sulit dilakukan.
